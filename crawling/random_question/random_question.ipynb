```python
rq_df = pd.read_csv("../data/random_question_240319.csv")
rq_df = rq_df.dropna(axis=0)
rq_df.info()
```

```python
seps = [".", "!", "~", "?"]
for i, row in rq_df.iterrows():
    if pd.isna(row["question"]):
        continue

    if "블로그씨" in row["question"]:
        for sep in seps:
            split_list = row["question"].strip().split(sep)
            split_list = [split.strip() for split in split_list]
            #print(split_list)
            if len(split_list) > 1 and len(split_list[-1])!=0:
                print(split_list[1])
                rq_df.loc[row["id"],"question"] = split_list[1]
                break
```

```python
rq_df_preprocessed = pd.read_csv("../data/random_question_preprocessed.csv")
rq_df_preprocessed.head(2)
```

```python
from datetime import datetime
```

```python
drop_idx = []
for i, row in rq_df_preprocessed.iterrows():
    if pd.isna(row["question"]):
        continue

    date = row["date"]
    print(date)
    try:
        date_obj = datetime.strptime(date, "%m월 %d일")
        if 33 < row["id"] and row["id"] < 184:
            date_obj = date_obj.replace(year=2024)
        else:
            date_obj = date_obj.replace(year=2025)
    except:
        drop_idx.append(i)
        continue

    rq_df_preprocessed.loc[row["id"],"date"] = date_obj.strftime("%Y-%m-%d")
```

```python
rq_df_preprocessed_final = rq_df_preprocessed[["question", "date"]]
rq_df_preprocessed_final=rq_df_preprocessed_final.dropna()
rq_df_preprocessed_final=rq_df_preprocessed_final.drop(drop_idx, axis=0)

rq_df_preprocessed_final=rq_df_preprocessed_final.sort_values(by="date")
rq_df_preprocessed_final=rq_df_preprocessed_final.reset_index(drop=True)
rq_df_preprocessed_final=rq_df_preprocessed_final.drop(0, axis=0)
rq_df_preprocessed_final=rq_df_preprocessed_final.reset_index(drop=True).reset_index(drop=False)
rq_df_preprocessed_final.head(3)
```

```python
for i, row in rq_df_preprocessed_final.iterrows():
    if pd.isna(row["question"]):
        continue

    q = row["question"]
    q = q.replace("\"", "")
    q = q.replace("\'", "")
    q = q.replace("\' ", "")
    q = q.strip()
    rq_df_preprocessed_final.loc[i,"question"] = q
    print(rq_df_preprocessed_final.loc[i,"question"])
```

```python
rq_df_preprocessed_final = rq_df_preprocessed_final.dropna()
rq_df_preprocessed_final.head(40)
```

```python
rq_df_preprocessed_final.to_csv("../data/random_question_preprocessed_final.csv", index=False)
```
